**Duration**

```python
def total_duration_in_minutes(x):
    
    if len(x.split(" "))==2:
        hrs,mins=x.split(" ")
        hmins=int(hrs.split("h")[0])*60
        mins=int(mins.split("m")[0])
        return hmins+mins
    
    elif len(x.split(" "))==1:
        mins=x.split(" ")[0]
        if "h" in mins:
            hmins=int(mins.split("h")[0])*60
            
            return hmins
        elif "m" in mins:
            hmins=int(mins.split("m")[0])
            return hmins
    
```

```python
df["Duration"]=df["Duration"].apply(total_duration_in_minutes)
```

**Date_of_Journey**

```python
def Date_mgmt(x):
    day,month,year=x.split("/")
    return day,month,year
```

```python
df['Day'], df['Month'], df['Year'] = zip(*df['Date_of_Journey'].map(Date_mgmt))
```

```python
#dropping the column
df.drop(labels='Date_of_Journey', axis=1,inplace=True)
```

**Dep_Time Arrival_Time**

```python
def timmings(x):
    if (re.search('[a-zA-Z]', x))!=None:
        return x.split(" ")[0]
    else:
        return x
```

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(timmings)
df["Dep_Time"]=df["Dep_Time"].apply(timmings)
```
