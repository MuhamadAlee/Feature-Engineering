```python
import pandas as pd
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df=pd.read_csv("Purchase.csv")
df
```

```python
df.info()
```

**Handling Missing Values**

```python
df.isna().sum()
```

```python
df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0], inplace=True)
df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
df
```

**Dropping User ID and Product ID**

```python
df.drop(columns=['User_ID', 'Product_ID'],axis=1,inplace=True)
```

```python
df
```

**Gender**

```python
df["Gender"]=le.fit_transform(df["Gender"])
df["Gender"]=df["Gender"].astype("int")
```

```python
df
```

**Age**

```python
df["Age"].unique()
```

```python
df["Age"]=le.fit_transform(df["Age"]).astype("int")
```

```python
df
```

**City_Category**

```python
df["City_Category"].unique()
```

```python
df["City_Category"]=le.fit_transform(df["City_Category"]).astype("int")
```

```python
df
```

**Stay_In_Current_City_Years**

```python
df["Stay_In_Current_City_Years"].unique()
```

```python
df["Stay_In_Current_City_Years"]=df["Stay_In_Current_City_Years"].map({"0":0,"1":1,"2":2,"3":3,"4+":4})
```
